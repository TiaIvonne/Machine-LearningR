```python
# Imports
import h2o
import matplotlib as plt
%matplotlib inline
from h2o.automl import H2OAutoML

```

```python
# Setting de los threads a utilizar
h2o.init(nthreads=8)
```

```python
# Lectura del archivo dengue
dataset = h2o.import_file("data_dengue1.csv")
```

```python
# C1 es el numero de fila
dataset.head()
```

```python
# Exploracion de la variable objetivo
dataset["varObjBin"].table()
```

```python
dataset[1] = dataset[1].asfactor()
```

```python
train = dataset
```

```python
# Se seleccionan las variables independientes y dependientes
predictors = ["humid", "humid90", "temp", "temp90", "h10pix", "h10pix90", 
    "trees", "trees90", "Xmin", "Xmax", "Ymin", "Ymax"]
response = "varObjBin"
```

```python
# Set autoML 
aml = H2OAutoML(max_runtime_secs = 900, 
                max_models = 25,  
                seed = 42, 
                project_name='classification',
                sort_metric = "AUC")
```

```python
# Entrenamiento del modelo
aml.train(x=predictors, y=response, training_frame=train)
```

```python
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```
